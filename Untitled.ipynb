```python
import random
import json
import pickle
import numpy as np

import nltk

from nltk.stem import WordNetLemmatizer

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD


nltk.download()
lemmatizer = WordNetLemmatizer()

characters_to_delete = ['?', '!', '.', ',', '@', '(', ')']
array_list_of_sentences_lemmatized = []
array_type_of_word_to_output = []
all_words_json = []

data = json.loads(open('data.json').read())

for datas_json in data['datas']:
    for question in datas_json['answers']:
        sentences_tokenized = nltk.word_tokenize(question) #jogarei joguei -> jogar
        array_list_of_sentences_lemmatized.extend(sentences_tokenized)
        all_words_json.append((sentences_tokenized, datas_json['key']))
        if datas_json['key'] not in array_type_of_word_to_output:
            array_type_of_word_to_output.append(datas_json['key'])

 
array_list_of_sentences_lemmatized = [lemmatizer.lemmatize(word) for word in array_list_of_sentences_lemmatized if word not in characters_to_delete]
array_list_of_sentences_lemmatized = sorted(set(array_list_of_sentences_lemmatized))

array_type_of_word_to_output = sorted(set(array_type_of_word_to_output))

pickle.dump(array_list_of_sentences_lemmatized, open('array_list_of_sentences_lemmatized.pkl', 'wb'))
pickle.dump(array_type_of_word_to_output, open('array_type_of_word_to_output.pkl', 'wb'))

zero_array = [0] * len(array_type_of_word_to_output)
training = []

for document in all_words_json:
    save_values_lemmatized = []
    sentence_answers_pattern = document[0]
    sentence_answers_pattern = [lemmatizer.lemmatize(word.lower()) for word in sentence_answers_pattern]
    for word in array_list_of_sentences_lemmatized:
        save_values_lemmatized.append(1) if word in sentence_answers_pattern else save_values_lemmatized.append(0)

    model_output_expected = list(zero_array)
    model_output_expected[array_type_of_word_to_output.index(document[1])] = 1
    training.append([save_values_lemmatized, model_output_expected])



random.shuffle(training)

training = np.array(training)

train_x = list(training[:, 0])
train_y = list(training[:, 1])

model = Sequential()
#256 neuronios, as entradas da funcao e relu
model.add(Dense(256, input_shape = (len(train_x[0]), ), activation = 'relu'))
#Esta função é usada para evitar overfitting em um modelo, definindo aleatoriamente uma taxa de fração
model.add(Dropout(0.5))
model.add(Dense(64, activation ='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation = 'softmax'))

sgd = SGD(lr=0.01, decay =1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics = ['accuracy'])

data_robot_voice = model.fit(np.array(train_x), np.array(train_y), epochs = 200, batch_size = 5, verbose = 1)
model.save('modelRobotVoice.h5', data_robot_voice)
print('model is created')
```
