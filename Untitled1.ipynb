```python
import random
import json
import pickle
import numpy as np
import nltk
from nltk.stem import WordNetLemmatizer
import pandas as pd
import pyttsx3 
import speech_recognition as sr
from tensorflow.keras.models import load_model

lemmatizer = WordNetLemmatizer()
data = json.loads(open('data.json').read())

sentences_lemmatized = pickle.load(open('array_list_of_sentences_lemmatized.pkl', 'rb'))
datas_pattern = pickle.load(open('array_type_of_word_to_output.pkl', 'rb'))


def model_loading():
    try:
        model = load_model('modelRobotVoice.h5') 
    except:
        raise Exception('model load fail for some reason')
    finally:
         return model

def get_word_lemmatized_from_bot(intents_list, intents_json):
    try:
        word_to_bot_say = ''
        key = intents_list[0]['intent']
        list_of_possible_words = intents_json['datas']
        for word in list_of_possible_words:
            if word['key'] == key:
                word_to_bot_say = random.choice(word['result'])
                break
        return word_to_bot_say
    except:
        print('fail to generate a random choice to bot')
        engine = pyttsx3.init() 
        engine.say("fail to random a choice to bot") 
        engine.runAndWait()
    finally:
        return word_to_bot_say


def word_tokenize(word):
    try: #worked working -> work
        word_lemmatizer = nltk.word_tokenize(word)
        word_lemmatizer = [lemmatizer.lemmatize(word) for word in word_lemmatizer]
    except IndexError:
        print('out of range to lemmatize the word')
    finally:
        return word_lemmatizer #played playing -> (play)


def audio_converter_bot_Word(word):
    try:
        engine = pyttsx3.init() 
        engine.say(word) 
        engine.runAndWait()
    except:
       print("An exception occurred while play the audio")
    
    
def array_of_words_lemmatized(word):
    try:
        valuesWordsLemmatized = [0] * (len(sentences_lemmatized))
        words_Lemmatisation = word_tokenize(word) # Lemmatize -> working, worked to work (example)
        for j in words_Lemmatisation:
            for i, word_lemmatized in enumerate(sentences_lemmatized):
                if word_lemmatized == j:
                    valuesWordsLemmatized[i] = 1
        return np.array(valuesWordsLemmatized)
    except:
        print('fail to Lemmatize the word')
    finally:
        return np.array(valuesWordsLemmatized)

def audio_converter_men_to_word():
    text = ""
    microphone = sr.Recognizer()
    with sr.Microphone() as source:
        microphone.adjust_for_ambient_noise(source)
        print('say something')
        audio = microphone.listen(source)
        try:
            text = microphone.recognize_google(audio, language="pt-BR")
            print('voce disse', text)
        except:
            engine = pyttsx3.init() 
            engine.say("eu nao entendi") 
            engine.runAndWait()
    return text


def possible_model_word(word, model):
    try:
        word_chance_results = []
        substantive = array_of_words_lemmatized(word)
        model_predict = model.predict(np.array([substantive]))[0]
        values = [[i, j] for i, j in enumerate(model_predict) if j > 0.24]
        values.sort(key=lambda x: x[1], reverse=True)
        for j in values:
            word_chance_results.append({'intent': datas_pattern[j[0]], 'probability': str(j[1])})
        return word_chance_results
    except:
        engine = pyttsx3.init() 
        engine.say("word predict fail") 
        engine.runAndWait()
        raise Exception("fail to predict word")
    finally:
        return word_chance_results

if __name__ == '__main__':
    model_loaded = model_loading()
    while True:
        print('ROBOT ONLINE')
        word_to_say = audio_converter_men_to_word()
        if word_to_say == 'exit':
            print('voce escolheu sair da conversa')
            engine = pyttsx3.init() 
            engine.say("obrigado pela conversa") 
            engine.runAndWait()
            break
        model_word_calculated = possible_model_word(word_to_say, model_loaded)
        word_calculated_to_bot_say = get_word_lemmatized_from_bot(model_word_calculated, data)
        audio_converter_bot_Word(word_calculated_to_bot_say)
        print(word_calculated_to_bot_say)
```
